**Мостиковая схема**

```python
import numpy as np
import matplotlib.pyplot as plt
import math

p = np.arange (0.5, 0.99, 0.001)
p_3 = 0.6

p_1 = p
p_2 = 0.5 *p
p_4 = p
p_5 = 0.5 *p
# Элемент 3 неисправен
pa1  = p_1*p_2
pa2  = p_4*p_5
pa = 1 - (1 - pa1)*(1 - pa2)
# Элемент 3 гарантированно исправен
pb1  = 1 - (1 -p_1)*(1 - p_4)
pb2  = 1 - (1 - p_2)*(1 - p_5)
pb = pb1*pb2
#
p_syst = (1 - p_3)*pa + p_3*pb
fig = plt.figure()   
plt.plot (p,p_syst, '-', label = 'p_syst')
plt.plot (p,pa, ':', label = 'pa')
plt.plot (p,pb, '-.', label = 'pb')
plt.title('Расчет надежности системы')  
plt.xlabel('Вероятность работоспособности элемента $p_3$ = 0.6')  
plt.ylabel('Надежность системы') 
plt.legend() 
plt.grid(True)
plt.show()

```
