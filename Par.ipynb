```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Вероятность работоспособности системы   задается формулой:

\begin {equation}
\label{e1_5_6a}  
\mathbb{P}= \mathbb{P}(\varphi(\textbf{X} = 1))  = \prod_{i=1}^{n}\mathbb{P}(x_i = 1) = \prod_{i=1}^{n}p_i.
\end {equation}


```python
T = 100
TT = []
k = 1
n = np.arange (2., 10.,1.)
for nn in n:
    k = k + 1/nn
    TT.append(T*k) 
```

В качестве иллюстрации, на  рисунке приведен график увеличения среднего времени наработки на отказ для системы, состоящей из набора параллельно соединенных элементов. В данном примере  среднее время наработки на отказ  элемента системы положено равным  $1/\lambda = 100$. На графике наглядно продемонстрировано,  что четырехкратное дублирование элемента приводит в чуть более чем духкратному увеличению времени наработки на отказ системы. 

А девять(!) параллельно подключенных элементов не приводят даже к трехкратному увеличению времени наработки на отказ системы. 

```python
plt.plot (n,TT)
plt.title('График роста времени наработки на отказ системы как функция роста числа параллельно подключаемых элементов')   
plt.xlabel('Число параллельно подключаемых элеменов')  
plt.ylabel('Время наработки на отказ системы')   
plt.grid(True)
plt.show()
```
