**Последовательное соединение элементов и подсистем**

```python
import numpy as np
import matplotlib.pyplot as plt
import math

p = np.arange (0.5, 0.99, 0.001)
p_1 = p
p_2 = 0.5 *p
pa1  = 1 - (1 - p_1)*(1 - p_2)
pa2  = 1 - (1 - p_1)*(1 - p_2)
pa = pa1*pa2
pb1  = p_1*p_2
pb2  = p_1*p_2
pb = 1 - (1 - pb1)*(1 - pb2)
fig = plt.figure()   # Отрисовка данных
plt.plot (p,pa, '-.')
plt.plot (p,pb, ':')
plt.title('Расчет надежности системы')  
plt.xlabel('Вероятность работоспособности элемента 1')  
plt.ylabel('Надежность системы')  
plt.grid(True)
plt.show()

```

**Параллельное соединение элементов и подсистем**

```python

p = np.arange (0.5, 0.99, 0.001)
p_1 = p
p_2 = 0.5 *p
pa1  = 1 - (1 - p_1)**2
pa2  = 1 - (1 - p_2)**2
pa = 1 - (1 - pa1)*(1 - pa2)
pb1  = 1 - (1 - p_1)*(1 - p_2)
pb2  = 1 - (1 - p_1)*(1 - p_2)
pb = 1 - (1 - pb1)*(1 - pb2)
fig = plt.figure()   # Отрисовка данных
plt.plot (p,pa, '-.')
plt.plot (p,pb, ':')
plt.title('Расчет надежности системы')  
plt.xlabel('Вероятность работоспособности элемента 1')  
plt.ylabel('Надежность системы')  
plt.grid(True)
plt.show()

```
